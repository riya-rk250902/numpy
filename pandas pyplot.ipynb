```python
# 17 july 23

import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
cars = pd.read_csv(r"C:\Users\Lenovo\csv data_files\csv file sample_data\doc_1\cars.csv")
cars.head()
```

```python
# plt = pyplot
cars.mpg.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("data of mpg")
plt.ylabel("number of cars")
plt.xlabel("mpg")
```

```python
# find the relation between the variables
# corr = correlation 
plt.figure(figsize=(20,10))
c= cars.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
